#### 3.3.4. Investigate the Day of 'Departure Date' Column

```python
sorted(dataset['Departure Day'].unique())
```

It can be seen that the range for days is 1 to 31 which makes sense. However, we have to consider month with total days less than 31 such as February, March, etc. Therefore, we will investigate the day range for each month.

```python
days_of_months = {'01': '31', '02': '28', '03':'31', '04':'30', '05':'31', '06':'30', '07':'31', '08':'31', '09':'30','10':'31','11':'30','12':'31'}
for month, day in days_of_months.items():
    total_rows = len(dataset[(dataset['Departure Month'] == month) & (dataset['Departure Day'] > day)])
    if total_rows > 0:
        print("Month", month, "has", total_rows, "wrong day values")
        display(dataset[(dataset['Departure Month'] == month) & (dataset['Departure Day'] > day)])
    else:
        print("Month", month,"has no wrong day value")
```

We found that there are some months with wrong day values. February has 27 rows, April has 35 rows and June has 38 rows. The total rows is 100. This is probably caused by the human error. The most reasonable solution is to change it into the last day of the month. We will fix it using this following step.

```python
days_of_months = {'01': '31', '02': '28', '03':'31', '04':'30', '05':'31', '06':'30', '07':'31', '08':'31', '09':'30','10':'31','11':'30','12':'31'}
for month, day in days_of_months.items():
    ids = dataset.index[(dataset['Departure Month'] == month) & (dataset['Departure Day'] > day)].tolist()
    for index in ids:
        dataset.loc[index,'Departure Day'] = day
```

```python
days_of_months = {'01': '31', '02': '28', '03':'31', '04':'30', '05':'31', '06':'30', '07':'31', '08':'31', '09':'30','10':'31','11':'30','12':'31'}
for month, day in days_of_months.items():
    total_rows = len(dataset[(dataset['Departure Month'] == month) & (dataset['Departure Day'] > day)])
    if total_rows > 0:
        print("Month : ", month)
        print("Total rows : ", total_rows)
        display(dataset[(dataset['Departure Month'] == month) & (dataset['Departure Day'] > day)])
    else:
        print("Month", month,"has no wrong day value")
```

We fixed and checked it again. There is no month with wrong day values again. It means the problem is fixed. The following step is to combine the 'Departure Year', 'Departure Month' and 'Depature Day' columns into 'Temp Departure Date' column.

```python
dataset['Temp Departure Date'] = dataset[['Departure Year', 'Departure Month', 'Departure Day']].apply(lambda x: '-'.join(x), axis=1)
```

In order to ensure that we have fixed it correctly, we will count the total different values between 'Departure Date' column and 'Temp Departure Date' column.

```python
(dataset['Departure Date'] != dataset['Temp Departure Date']).sum()
```

It can be seen that there are 213 different values. This makes sense because we fixed 113 wrong format dates and 100 wrong day values. Now we will replace the original dates with the correct dates.

```python
dataset['Departure Date'] = dataset['Temp Departure Date'].copy()
```

In order to ensure that we do in the right way, we will try to calculate the distance with the destination and compare them with the given 'Journey Distance'

```python
# Function for calculating the distance between 2 latitudes and longitudes
def calculate_radian (lat1, lon1, lat2, lon2):
    R = 6378
    dlon = radians(lon2) - radians(lon1)
    dlat = radians(lat2) - radians(lat1)

    a = sin(dlat / 2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return round(distance,3)
```

It can be seen that there are 3 types of drone. Then what distinguishes them. From the given specification, we know that each type of drone has three phases of flight (namely takeOff, onRoute, and Landing). The drone may have different speeds at different phases. takeOff and Landing phases only take five minutes. We will try to visualize it.

```python
sns.lmplot( x="Journey Distance", y="Travel Time", data=dataset, fit_reg=False, 
           hue='Drone Type', legend=True, size =5)
ax = plt.gca()
ax.set_title("Journey Distance compared to Travel Time")
```

Now we will calculate the speed for takeOff and Landing and also onRoute by using Drone Type, Travel Time and Journey Distance. We will use the procedure like this.
1. Exclude data with null value
2. Split data based on the drone type
3. For each drone type, split the data into 2 groups
4. For each group, calculate the average of travel time and the average of journey distance, then form this equations
    - Assume that takeoff = x, Landing = y, and onRoute = z
    - Assume xy = (5x + 5y), the traveled distance during takeoff and landing
    - Assume z = OnRoute speed in minute
    - Assume Avg_TT = Average of Travel Time
    - Assume Avg_JD = Average of Journey Distance
    - Therefore, we will get:
        * xy1 + (Avg_TT1 - 10)z1 = Avg_JD1
        * xy1 + (Avg_TT2 - 10)z1 = Avg_JD2
    - After, we get the xy and z value, we calculate the new travel time
    - Exclude the data with huge difference between given travel time and the new travel time. We select the data based on IQR. After that, repeat step number 4.
5. We will stop the looping until the new travel time fits 100% with the given travel time.

```python
dataset.describe()
```

However, in order to get the best result, we will exclude row with 'Drone Type' null and 'Travel Time' null.

```python
dataset['Condition']='True'
dataset['Condition'][dataset['Drone Type'].isnull()]='Drone Type Null'
dataset['Condition'][dataset['Travel Time'].isnull()]='Travel Time Null'
```

Copy dataset in order to support the calculation process.

```python
df = dataset[dataset['Condition']=='True'].copy()
```

We will calculate the xy and z values for Drone Type 1.0.

```python
df1 = df[['Id','Drone Type','Journey Distance', 'Travel Time','Condition']][(df['Drone Type']==1.0) & (df['Condition']=='True')]

fitness=0

# We will loop until the fitness equal to 1
while fitness!=1:
    df1 = df1[df1['Condition']=='True'].copy()
    avg_dist_list1 = []
    avg_time_list1 = []

    # Function for splitting the rows into 2 group
    def iter_by_group(df, column, num_groups):
        groups = []
        for i, group in df.groupby(column):
            groups.append(group)
            if len(groups) == num_groups:
                yield pd.concat(groups)
                groups = []
        if groups:
            yield pd.concat(groups)
    # Calculate the average of travel time and the average of journey distance
    for group in iter_by_group(df1, 'Id', ceil(len(df1)/2)):
        avg_time_list1.append(group['Travel Time'].mean())
        avg_dist_list1.append(group['Journey Distance'].mean())

    # Calculate the OnRoute Speed in Minute
    z1 = (avg_dist_list1[0]-avg_dist_list1[1])/((avg_time_list1[0]-10)-(avg_time_list1[1]-10))

    # Calculate the traveled distance during TakeOff and Landing
    xy1 = avg_dist_list1[0] - (avg_time_list1[0]-10)*z1

    # Calculate new travel time
    df1['Travel Time D1'] = ((df1['Journey Distance']-(xy1))/(z1))+10
    df1['Abs D1'] = abs(df1['Travel Time']-df1['Travel Time D1'])
    
    # Calculate the IQR
    q75, q25 = np.percentile(df1['Abs D1'], [75 ,25])
    iqr = q75 - q25
 
    max_abs = q75 + (iqr*1.5)
    
    fitness = len(df1[df1['Abs D1']<max_abs])/len(df1)
    
    # Exclude the row with distance larger than IQR
    df1['Condition'][df1['Abs D1']>max_abs]='False'
print("XY1 :",xy1)
print("Z1 :", z1)
```

We have fixed the format of 'Delivery Time' in the previous section. In this part, we will try to check the correctness of the value in 'Delivery Time' column. We can do it by minus 'Delivery Time' and 'Departure Time' and compare the result with the given 'Travel Time'

```python
# Function for calculating the travel time
def calculate_time(hour1, minute1, second1, hour2, minute2, second2):
    resulth = hour2 - hour1
    resultm = minute2 - minute1
    results = second2 - second1
    if (results <0):
        results +=60
        resultm-=1
    if(resultm<0):
        resultm+=60
        resulth-=1
    if(resulth<0):
        resulth+=24
    return resulth*60 + resultm + results/60
```

It can be seen that if we calculate new travel time using 'Delivery Time' - 'Departure Time'. The result will be same with the result if we calculate using drone speed and journey distance. Therefore, we will impute the travel time using this scenario.

```python
# Function for fill the travel time
def fillna_traveltime(drone, tempTT, TT1, TT2, TT3):
    if ((abs(tempTT-TT1)<abs(tempTT-TT2)) & (abs(tempTT-TT1)<abs(tempTT-TT3))):
        if drone==1.0:
            return round(tempTT,2)
    if ((abs(tempTT-TT2)<abs(tempTT-TT1)) & (abs(tempTT-TT2)<abs(tempTT-TT3))):
        if drone==2.0:
            return round(tempTT,2)
    if ((abs(tempTT-TT3)<abs(tempTT-TT1)) & (abs(tempTT-TT3)<abs(tempTT-TT2))):
        if drone==3.0:
            return round(tempTT,2)
```

From the result we got R-squared = 0.9923097447370298
R-squared is a statistical measure of how close the data are to the fitted regression line. It is also known as the coefficient of determination, or the coefficient of multiple determination for multiple regression.

The definition of R-squared is fairly straight-forward; it is the percentage of the response variable variation that is explained by a linear model. Or:

R-squared = Explained variation / Total variation

R-squared is always between 0 and 100%:

0% indicates that the model explains none of the variability of the response data around its mean. 100% indicates that the model explains all the variability of the response data around its mean.

In addtion, we also got coefficients = 38.30641918, 32.77251895, 15.18581525, 0.0893853, which means

We also got intercept = 26.142748236000457

Therefore, we can form a linear model

\begin{equation}
y_i = 38.30641918 A_i + 32.77251895 B_i + 15.18581525 C_i + 0.0893853 D_i + E
\end{equation}

where:
A = Drone Type, 
B = Post Type, 
C = Cluster Time (1:Morning;2:Afternoon;3:Night), 
D = Journey Distance, 
E = Intercept
